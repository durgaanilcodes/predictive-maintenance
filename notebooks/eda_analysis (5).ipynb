# Correlation Heatmap

```python
numeric_df = df.select_dtypes(include=['int64','float'])
numeric_df.head()
```

```python
correlation = numeric_df.corr()
correlation
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm"
)
plt.title("Feature Correlation Heatmap")
plt.show()
```

## Heatmap Analysis
- Correlation heatmap shows relationships between numerical features.
- Strong positive and negative relationships can be identified.
- This helps in selecting important features for machine learning.

# Feature Relationship Analysis

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    x='Machine failure',
    y='Air temperature [K]',
    data=df
)
plt.title("Air Temperature vs Machine Failure")
plt.show()
```

## Observation

- Air temperature distribution was compared with machine failure.
- The chart helps identify whether temperature changes affect failure cases.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='Rotational speed [rpm]',
    y='Torque [Nm]',
    hue='Machine failure',
    data=df
)
plt.title("Torque vs Rotational Speed")
plt.show()
```

## Observation

- Torque and rotational speed relationships was visualized.
- Failure points can be compared with operating conditions.
- These relationships can help in predictive maintenance analysis.
