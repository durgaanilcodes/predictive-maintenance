```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("ai4i2020.csv")

print(df.head())
```

```python
print("Rows and Columns:")
print(df.shape)
```

```python
print(df.columns)
```

```python
#Feature 1:Temperature Difference
df["temp_difference"] = (
    df["Process temperature [K]"]
    - df["Air temperature [K]"]
)

print(df["temp_difference"].head())
```

```python
#Feature 2: Power Feature
df["power_feature"] = (
    df["Torque [Nm]"]
    *
    df["Rotational speed [rpm]"]
)

print(df["power_feature"].head())
```

```python
#Feature 3: Wear-Speed Ratio
df["wear_speed_ratio"] = (
    df["Tool wear [min]"]
    /
    df["Rotational speed [rpm]"]
)

print(df["wear_speed_ratio"].head())
```

```python
#Feature 4: Wear-Torque
df["wear_torque"] = (
    df["Tool wear [min]"]
    *
    df["Torque [Nm]"]
)

print(df["wear_torque"].head())
```
