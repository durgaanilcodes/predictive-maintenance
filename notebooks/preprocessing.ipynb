```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("ai4i2020.csv")

print(df.head())
```

```python
print("Rows and Columns:")
print(df.shape)
```

```python
print(df.columns)
```

```python
#Feature 1:Temperature Difference
df["temp_difference"] = (
    df["Process temperature [K]"]
    - df["Air temperature [K]"]
)

print(df["temp_difference"].head())
```

```python
#Feature 2: Power Feature
df["power_feature"] = (
    df["Torque [Nm]"]
    *
    df["Rotational speed [rpm]"]
)

print(df["power_feature"].head())
```

```python
#Feature 3: Wear-Speed Ratio
df["wear_speed_ratio"] = (
    df["Tool wear [min]"]
    /
    df["Rotational speed [rpm]"]
)

print(df["wear_speed_ratio"].head())
```

```python
#Feature 4: Wear-Torque
df["wear_torque"] = (
    df["Tool wear [min]"]
    *
    df["Torque [Nm]"]
)

print(df["wear_torque"].head())
```

```python
print(df.head())
```

```python
# ===================================
# TASK 2: STATISTICAL FEATURES
# ===================================
```

```python
#Create Torque Z-Score
df["torque_zscore"] = (
    df["Torque [Nm]"]
    - df["Torque [Nm]"].mean()
) / df["Torque [Nm]"].std()

print(df["torque_zscore"].head())
```

```python
#Create Speed Z-Score
df["speed_zscore"] = (
    df["Rotational speed [rpm]"]
    - df["Rotational speed [rpm]"].mean()
) / df["Rotational speed [rpm]"].std()

print(df["speed_zscore"].head())
```

```python
#Create Temperature Z-Score
df["temp_zscore"] = (
    df["Air temperature [K]"]
    - df["Air temperature [K]"].mean()
) / df["Air temperature [K]"].std()

print(df["temp_zscore"].head())
```

```python
#Verify New Features
print(df.columns)
```

```python
#View Updated Dataset
print(df.head())
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
#Encode the Type Column
encoder = LabelEncoder()

df["Type"] = encoder.fit_transform(df["Type"])

print(df["Type"].head())
```

```python
#Remove Unnecessary Columns
df = df.drop(
    columns=[
        "UDI",
        "Product ID"
    ]
)

print(df.columns)
```

```python
#Scale Numerical Features
scaler = StandardScaler()

cols = [
    "Air temperature [K]",
    "Process temperature [K]",
    "Rotational speed [rpm]",
    "Torque [Nm]",
    "Tool wear [min]"
]

df[cols] = scaler.fit_transform(df[cols])
```
