```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
numeric_df.hist(figsize=(15,10))
plt.tight_layout()
plt.show()
```
