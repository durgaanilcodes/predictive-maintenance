Correlation Heatmap Analysis

```python
numeric_df = df.select_dtypes(include=['int64','float64'])
numeric_df.head()
```

```python
correlation = numeric_df.corr()
correlation
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(
    correlation,
    annot=True,
    cmap="coolwarm"
)
plt.title("Feature Correlation Heatmap")
plt.show()
```

## Observation

- Correlation heatmap shows relationships between numerical features.
- Values close to +1 show strong positive relationship.
- Values close to -1 show negative relationship.
- This analysis helps identify important features for prediction.
