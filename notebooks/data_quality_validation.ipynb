```python
# ===================================
# ACTIVITY 2
# Feature Distribution and Class Imbalance
# ===================================
```

```python
import matplotlib.pyplot as plt
```

```python
print(df["Machine failure"].value_counts())
```

```python
df["Machine failure"].value_counts().plot(kind="bar")

plt.title("Machine Failure Distribution")
plt.xlabel("Machine Failure")
plt.ylabel("Count")

plt.show()
```

```python
df["Air temperature [K]"].hist()

plt.title("Air Temperature Distribution")
plt.xlabel("Air Temperature")

plt.show()
```

```python
df["Process temperature [K]"].hist()

plt.title("Process Temperature Distribution")
plt.xlabel("Process Temperature")

plt.show()
```

```python
df["Rotational speed [rpm]"].hist()

plt.title("Rotational Speed Distribution")
plt.xlabel("Rotational Speed")

plt.show()
```

```python
df["Torque [Nm]"].hist()

plt.title("Torque Distribution")
plt.xlabel("Torque")

plt.show()
```

```python
df["Tool wear [min]"].hist()

plt.title("Tool Wear Distribution")
plt.xlabel("Tool Wear")

plt.show()
```

Feature Distribution Analysis

Observations:

1. Air temperature follows a normal distribution.
2. Process temperature follows a normal distribution.
3. Rotational speed shows moderate variation.
4. Torque values are normally distributed.
5. Tool wear values are spread across the range.
6. Machine failure is imbalanced, with significantly fewer failure records.
