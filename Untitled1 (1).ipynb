```python
import pandas as pd

df = pd.read_csv('ai4i2020.csv')
```

```python
df.head()
```

```python
import numpy as np

df['Ambient_Temperature'] = np.random.randint(20, 40, len(df))
df['Load_Density'] = np.random.randint(50, 100, len(df))
```

```python
df['Temp_Difference'] = (
    df['Process temperature [K]']
    - df['Air temperature [K]']
)

df['Power_Index'] = (
    df['Rotational speed [rpm]']
    * df['Torque [Nm]']
)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm')
plt.title("Feature Correlation Matrix")
plt.show()
```

```python
print(df[['Machine failure',
          'Ambient_Temperature',
          'Load_Density',
          'Temp_Difference',
          'Power_Index']].corr())
```
